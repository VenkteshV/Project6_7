```python

def intersection(lst1, lst2):  
  list2=[]        
  s= set(lst1).intersection(set(lst2))
  for i in s:
    list2.append(i)
  return list2
```

```python
txt_data = []
txt_data2 = []
txt_taxonomy=[]
for i in range(len(list_tagrec)):
  for k in range(len(list_tagrec2)):  
    common = intersection(list_tagrec[i], list_tagrec2[k])
    if(len(common)>0):
      txt_data.append(data['text'][i])
      txt_data2.append(data2['text'][k])
      txt_taxonomy.append(common)
print(len(list_tagrec))
print(len(list_tagrec2))
print(len(txt_data))
print(len(txt_data2))
print(len(txt_taxonomy))
```

```python
import textdistance
def get_closest(sent1, list2):  # sentence compared w list and jaccard score is calculated
  l_list = []
  for i in list2:
    l_list.append(textdistance.jaccard(sent1, i))
  l_list = np.array(l_list)
  max = l_list.argmax()
  return list2[max]

  
```

```python
def split_text(text):
  grouped = []
  ex_split = text.split(".")
  l = len(ex_split) % 3
  ll = len(ex_split) - l
  for i in range(0,ll, 3):
    tupl = (ex_split[i], ex_split[i+1],ex_split[i+2])
    t = "".join(tupl)
    grouped.append(t)
  return grouped
```
